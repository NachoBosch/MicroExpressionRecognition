# Elegimos Random im√°genes de No MicroExpresiones

```python
import os
import numpy as np
import pandas as pd
```

```python
dir = r"C:\Doctorado\Neurociencia\CASME\CASME-II-Binary"
imgs = os.listdir(f"{dir}/NE_original")
print(imgs)
```

```python
print(len(imgs))
```

```python
imgs_me = os.listdir(f"{dir}/ME")
print(len(imgs_me))
```

```python
imgs_no_me = np.random.choice(imgs,16159,replace=False)
print(imgs_no_me)
```

```python
import shutil

for i, img_nombre in enumerate(imgs_no_me):
    # Ruta de origen (imagen original)
    ruta_origen = os.path.join(f"{dir}/NE_original", img_nombre)
    # Ruta de destino (nueva carpeta)
    ruta_destino = os.path.join(f"{dir}/NE", img_nombre)
    # Copiar el archivo
    shutil.copy2(ruta_origen, ruta_destino)
    
    # Mostrar progreso cada 1000 im√°genes
    if (i + 1) % 1000 == 0:
        print(f"Procesadas {i + 1} im√°genes...")
```
