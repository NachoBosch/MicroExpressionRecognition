```python
import matplotlib.pyplot as plt
import pandas as pd

# Datos
data = {
    'Subset': ['train', 'train', 'val', 'val'],
    'Class': ['ME', 'Others', 'ME', 'Others'],
    'n_images': [8477, 5093, 2120, 1274]
}

df = pd.DataFrame(data)

# Gráfico
plt.figure(figsize=(8, 6))
colors = ['#4c72b0', '#dd8452']  # Colores para ME y Others

# Usar agrupamiento por clase
for i, cls in enumerate(df['Class'].unique()):
    subset = df[df['Class'] == cls]
    plt.bar(subset['Subset'], subset['n_images'], bottom=df[df['Class'] == df['Class'].unique()[i - 1]]['n_images'] if i > 0 else None, label=cls, color=colors[i])

# Etiquetas
plt.title('Balance de Imágenes de CASME II')
plt.xlabel('Conjunto')
plt.ylabel('Imágenes')
plt.legend(title='Clase')
plt.tight_layout()
plt.savefig('balance-binary-casme-2.png')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Subset', y='n_images', hue='Class', palette='viridis')
plt.title('Balance de Imágenes de CASME II')
plt.xlabel('Conjunto')
plt.ylabel('Imágenes')
plt.tight_layout()
plt.savefig('balance-binary-casme-2.png')
plt.show()
```
